```python
from mplsoccer import Pitch,VerticalPitch
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
```

```python
data = pd.read_csv('Match1_SPC.csv')
```

```python
data
```

```python
data['X'] = data['X']*1.2
data['Y'] = data['Y']*.8
data['X2'] = data['X2']*1.2
data['Y2'] = data['Y2']*.8
```

```python
cornerL = data[data['Y'] == 0.0 ]
cornerR = data[data['Y'] == 80.0 ]
```

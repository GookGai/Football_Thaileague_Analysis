```python
from mplsoccer import Pitch,VerticalPitch
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
```

```python
data = pd.read_csv('Match1_CFC.csv')
```

```python
data
```

```python
data['X'] = data['X']*1.2
data['Y'] = data['Y']*.8
data['X2'] = data['X2']*1.2
data['Y2'] = data['Y2']*.8
```

```python
data = data[data['Team'] == 'Away' ]
```

```python
data
```

```python
pitch = VerticalPitch(pitch_length=100, pitch_width=100,
        half = True,axis=True, label=True,pitch_color = '#538053')
fig,ax =pitch.draw(figsize =(20,10))

plt.scatter(data['Y2'],data['X2'],s = 300)
plt.title('All Corner Vs SPC',size = 20)
```

```python
pitch = VerticalPitch(pitch_length=100, pitch_width=100,
        half = True,axis=True,pitch_color = '#538053')
fig,ax =pitch.draw(figsize =(20,10))

for i in range(5):
    plt.scatter(data['Y2']+i*0.1,data['X2']+i*0.1,alpha=0.5,s = 250)
for i in range(5):
    plt.scatter(data['Y2']-i*0.1,data['X2']+i*0.1,alpha=0.5,s = 250)
for i in range(5):
    plt.scatter(data['Y2']+i*0.1,data['X2']-i*0.1,alpha=0.5,s = 250)
for i in range(5):
    plt.scatter(data['Y2']-i*0.1,data['X2']-i*0.1,alpha=0.5,s = 250)

plt.title('Match 1 Chiang Rai United  ',size = 20)
```

```python
pitch = VerticalPitch(pitch_length=100, pitch_width=100,
        half = True,axis=True,pitch_color = '#538053')
fig,ax =pitch.draw(figsize =(20,10))

for i in range(5):
    plt.scatter(data['Y2'],data['X2'],alpha=0.5+i*0.1,s = 3000-i*500)

plt.title('Match 1 Chiang Rai United  ',size = 20)
```

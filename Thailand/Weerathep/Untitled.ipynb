```python
from mplsoccer import Pitch,VerticalPitch
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
```

```python
data = pd.read_csv('firstHalf.csv')
data2 = pd.read_csv('secoundHalf.csv')
```

```python
data
```

```python
data['X'] = data['X']*1.2
data['Y'] = data['Y']*.8
data['X2'] = data['X2']*1.2
data['Y2'] = data['Y2']*.8

data2['X'] = data2['X']*1.2
data2['Y'] = data2['Y']*.8
data2['X2'] = data2['X2']*1.2
data2['Y2'] = data2['Y2']*.8
```

```python
data['X'] = 120-data['X']
data['Y'] = 80-data['Y']
data['X2'] = 120-data['X2']
data['Y2'] = 80-data['Y2']
```

```python
dataPass = data[(data['Event'] == 'Pass') |  (data['Event'] == 'Through') |
                         (data['Event'] == 'PassFail') | (data['Event'] == 'Cross')]
dataPass = dataPass.reset_index()
pDrtx = dataPass['X']
pDrty = dataPass['Y']
pDesx = dataPass['X2']
pDesy = dataPass['Y2']
pSty = dataPass['Event']
```

```python
dataPass15 = data[((data['Event'] == 'Pass') |  (data['Event'] == 'Through') |
                         (data['Event'] == 'PassFail') | (data['Event'] == 'Cross')) & ((data['Mins'] >0)&(data['Mins'] <=  15))]
dataPass30 = data[((data['Event'] == 'Pass') |  (data['Event'] == 'Through') |
                          (data['Event'] == 'PassFail') | (data['Event'] == 'Cross'))& ((data['Mins'] >15)&(data['Mins'] <=  30))]
dataPass45 = data[((data['Event'] == 'Pass') |  (data['Event'] == 'Through') |
                          (data['Event'] == 'PassFail') | (data['Event'] == 'Cross'))& ((data['Mins'] >30)&(data['Mins'] <=  46))]

dataPass15 = dataPass15.reset_index()
pDrtx15 = dataPass15['X']
pDrty15 = dataPass15['Y']
pDesx15 = dataPass15['X2']
pDesy15 = dataPass15['Y2']
pSty15 = dataPass15['Event']

dataPass30 = dataPass30.reset_index()
pDrtx30 = dataPass30['X']
pDrty30 = dataPass30['Y']
pDesx30 = dataPass30['X2']
pDesy30 = dataPass30['Y2']
pSty30 = dataPass30['Event']

dataPass45 = dataPass45.reset_index()
pDrtx45 = dataPass45['X']
pDrty45 = dataPass45['Y']
pDesx45 = dataPass45['X2']
pDesy45 = dataPass45['Y2']
pSty45 = dataPass45['Event']
```

```python
dataPass2 = data2[(data2['Event'] == 'Pass') |  (data2['Event'] == 'Through') |
                         (data2['Event'] == 'PassFail') | (data2['Event'] == 'Cross')]
dataPass2 = dataPass2.reset_index()
pDrtx2 = dataPass2['X']
pDrty2 = dataPass2['Y']
pDesx2 = dataPass2['X2']
pDesy2 = dataPass2['Y2']
pSty2 = dataPass2['Event']
```

```python
dataPass15h = data2[((data2['Event'] == 'Pass') |  (data2['Event'] == 'Through') |
                         (data2['Event'] == 'PassFail') | (data2['Event'] == 'Cross')) & ((data2['Mins'] >44)&(data2['Mins'] <=  60))]
dataPass30h = data2[((data2['Event'] == 'Pass') |  (data2['Event'] == 'Through') |
                          (data2['Event'] == 'PassFail') | (data2['Event'] == 'Cross'))& ((data2['Mins'] >60)&(data2['Mins'] <=  75))]
dataPass45h = data2[((data2['Event'] == 'Pass') |  (data2['Event'] == 'Through') |
                          (data2['Event'] == 'PassFail') | (data2['Event'] == 'Cross'))& ((data2['Mins'] >75)&(data2['Mins'] <=  90))]

dataPass15h = dataPass15h.reset_index()
pDrtx15h = dataPass15h['X']
pDrty15h = dataPass15h['Y']
pDesx15h = dataPass15h['X2']
pDesy15h = dataPass15h['Y2']
pSty15h = dataPass15h['Event']

dataPass30h = dataPass30h.reset_index()
pDrtx30h = dataPass30h['X']
pDrty30h = dataPass30h['Y']
pDesx30h = dataPass30h['X2']
pDesy30h = dataPass30h['Y2']
pSty30h = dataPass30h['Event']

dataPass45h = dataPass45h.reset_index()
pDrtx45h = dataPass45h['X']
pDrty45h = dataPass45h['Y']
pDesx45h = dataPass45h['X2']
pDesy45h = dataPass45h['Y2']
pSty45h = dataPass45h['Event']
```

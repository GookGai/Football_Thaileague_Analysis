```python
from mplsoccer import Pitch,VerticalPitch
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
```

```python
data = pd.read_csv('events (2).csv')
```

```python
data.head()
```

```python
data['X'] = data['X']*1.2
data['Y'] = data['Y']*.8
data['X2'] = data['X2']*1.2
data['Y2'] = data['Y2']*.8
```

```python
dataShot = data[(data['Event'] == 'ShotOnTarget') |  (data['Event'] == 'ShotOffTarget') |
                         (data['Event'] == 'ShotGetGoal') | (data['Event'] == 'ShotBlock')]
dataShot= dataShot.reset_index()
```

```python
dataShot
```

```python
pitch = VerticalPitch(pitch_length=100, pitch_width=100,
              axis=True, label=True,pitch_color = '#538053')
fig,ax =pitch.draw(figsize =(20,10))
    
plt.scatter(dataShot['Y'],dataShot['X'],c='#BE2653',s = 700,marker = "*")
```

```python
data['X'] = 120-data['X']
data['Y'] = 80-data['Y']
data['X2'] = 120-data['X2']
data['Y2'] = 80-data['Y2']
```

```python
data.head()
```

```python
dataShot = data[(data['Event'] == 'ShotOnTarget') |  (data['Event'] == 'ShotOffTarget') |
                         (data['Event'] == 'ShotGetGoal') | (data['Event'] == 'ShotBlock')]
dataShot= dataShot.reset_index()
```

```python
pitch = VerticalPitch(pitch_length=100, pitch_width=100,
              axis=True, label=True,pitch_color = '#538053',half = True)
fig,ax =pitch.draw(figsize =(20,10))
    
plt.scatter(dataShot['Y'],dataShot['X'],c='#BE2653',s = 500,marker = "*")
```

```python
dataPass = data[(data['Event'] == 'Pass') |  (data['Event'] == 'Through') |
                         (data['Event'] == 'PassFail') | (data['Event'] == 'Cross')]
```

```python
dataPass
```

```python
dataPass = dataPass.reset_index()
dataPass
```

```python
pDrtx = dataPass['X']
pDrty = dataPass['Y']
pDesx = dataPass['X2']
pDesy = dataPass['Y2']
pSty = dataPass['Event']
ailung = 0
unNar = 0
f = 0
```

```python
pitch = VerticalPitch(pitch_length=100, pitch_width=100,
              axis=True,pitch_color = '#538053',half=False)
fig,ax =pitch.draw(figsize =(40,20))


plt.scatter(dataPass['Y'],dataPass['X'],c='#60D8F6',s = 250 ,marker = '*')

for i in range(len(pSty)):
    if(pSty[i] == 'Cross'):
        plt.arrow(pDrty[i],pDrtx[i],pDesy[i]-pDrty[i],pDesx[i]-pDrtx[i], color = '#FF69B4',linewidth =3,head_width = 0.3 ,
                  length_includes_head = 'True')
    if(pSty[i] == 'Through'):
        plt.arrow(pDrty[i],pDrtx[i],pDesy[i]-pDrty[i],pDesx[i]-pDrtx[i], color = '#000000',linewidth =3,head_width = 0.3 ,
                  length_includes_head = 'True')
    if(pSty[i] == 'Pass'):
         plt.arrow(pDrty[i],pDrtx[i],pDesy[i]-pDrty[i],pDesx[i]-pDrtx[i],color ='#F6BF60',linewidth =3,head_width = 0.3 ,
                  length_includes_head = 'True')
    if(pSty[i] == 'PassFail'):
         plt.arrow(pDrty[i],pDrtx[i],pDesy[i]-pDrty[i],pDesx[i]-pDrtx[i],color ='#F0261E',linewidth =3,head_width = 0.3 ,
                  length_includes_head = 'True')
for j in range(len(pSty)):
    if(pDrtx[j] > pDesx[j]):
        ailung += 1
    if(pDrtx[j] < pDesx[j]):
        unNar += 1
for j in range(len(pSty)):
    if(pSty[j] == 'PassFail'):
        f +=1
plt.title('Thaliand Passing Direction In Philippines Match ',fontsize = 40)
```

```python
z = unNar,ailung
colors = ( "#6FC7F0" ,"#F07F8A")
plt.title("Team Passing Direction Style")
plt.pie(z ,startangle = 90, autopct='%1.2f%%', shadow=True,colors = colors)
plt.show()
```

```python
k = len(pSty)-f,f
colors = ( "#97F0B0","#F07F8A")
plt.title("Passing Accuracy")
plt.pie(k ,startangle = 90, autopct='%1.2f%%', shadow=True,colors = colors)
plt.show()
```

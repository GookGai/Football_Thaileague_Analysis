```python
from mplsoccer import Pitch,VerticalPitch
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
datashot = pd.read_csv('shot.csv')
datadef = pd.read_csv('def.csv')
datapass = pd.read_csv('pass.csv')
```

```python
datashot
```

```python
datadef
```

```python
datapass
```

```python
datashot['X'] = datashot['X']*1.2
datashot['Y'] = datashot['Y']*.8
datapass['X'] = datapass['X']*1.2
datapass['Y'] = datapass['Y']*.8
datapass['X2'] = datapass['X2']*1.2
datapass['Y2'] = datapass['Y2']*.8
datadef['X'] = datadef['X']*1.2
datadef['Y'] = datadef['Y']*.8
```

```python
datapass
```

```python
sonL = datashot[datashot['Event'] == 'ShotOnTargetLeft' ]
sofL = datashot[datashot['Event'] == 'ShotOffTargetLeft' ]
sggL = datashot[datashot['Event'] == 'ShotGetGoalLeft' ]
#
sonR = datashot[datashot['Event'] == 'ShotOnTargetRight' ]
sofR = datashot[datashot['Event'] == 'ShotOffTargetRight' ]
sggR = datashot[datashot['Event'] == 'ShotGetGoalRight' ]
#
sonH = datashot[datashot['Event'] == 'ShotOnTargetHead' ]
sofH = datashot[datashot['Event'] == 'ShotOffTargetHead' ]
sggH = datashot[datashot['Event'] == 'ShotGetGoalHead' ]
#
sbL = datashot[datashot['Event'] == 'ShotBlockLeft' ]
sbR = datashot[datashot['Event'] == 'ShotBlockRight' ]
sbH = datashot[datashot['Event'] == 'ShotBlockHead' ]
```

```python
shX = datashot['X']
shY = datashot['Y']
shName = datashot['Player']
```

```python
pCross = datapass[datapass['Event'] == 'cross' ]
pThrough = datapass[datapass['Event'] == 'through' ]
pWin = datapass[datapass['Event'] == 'win1-1' ]
```

```python
pDrtx = datapass['X']
pDrty = datapass['Y']
pDesx = datapass['X2']
pDesy = datapass['Y2']
pSty = datapass['Event']
pNamep = datapass['Player']
```

```python
pSty
```

```python
dfClear = datadef[datadef['Event'] == 'clear' ]
dfItc = datadef[datadef['Event'] == 'intercept' ]
dfGb= datadef[datadef['Event'] == 'getball' ]
dfBlk = datadef[datadef['Event'] == 'block' ]
dfYellow = datadef[datadef['Event'] == 'yellow' ]
dfRed = datadef[datadef['Event'] == 'red' ]
dfLose = datadef[datadef['Event'] == 'lose1-1' ]
dfFoul = datadef[datadef['Event'] == 'foul' ]
```

```python
dfClear
```

```python
pitch = VerticalPitch(pitch_length=100, pitch_width=100,
              axis=True, label=True,pitch_color = '#538053',half=True)
fig,ax =pitch.draw(figsize =(20,10))
    
plt.scatter(sonL['Y'],sonL['X'],c='#BE2AF3',s = 250,marker = "^")
plt.scatter(sofL['Y'],sofL['X'],c='#BE2AF3',s = 250,marker = "h")
plt.scatter(sggL['Y'],sggL['X'],c='#BE2AF3',s = 250 ,marker = '*')
plt.scatter(sbL['Y'],sbL['X'],c='#BE2AF3',s = 250,marker = "X")

plt.scatter(sonR['Y'],sonR['X'],c='#60D8F6',s = 250,marker = "^")
plt.scatter(sofR['Y'],sofR['X'],c='#60D8F6',s = 250,marker = "h" )
plt.scatter(sggR['Y'],sggR['X'],c='#60D8F6',s = 250 ,marker = '*')
plt.scatter(sbR['Y'],sbR['X'],c='#60D8F6',s = 250 ,marker = "X")

plt.scatter(sonH['Y'],sonH['X'],c='#F6BF60',s = 250,marker = "^")
plt.scatter(sofH['Y'],sofH['X'],c='#F6BF60',s = 250,marker = "h")
plt.scatter(sggH['Y'],sggH['X'],c='#F6BF60',s = 250 ,marker = '*')
plt.scatter(sbH['Y'],sbH['X'],c='#F6BF60',s = 250,marker = "X")


#for i in range(len(shX)):
 #   plt.text(shY[i]-1,shX[i]-2,shName[i])
```

```python
pitch = VerticalPitch(pitch_length=100, pitch_width=100,
              axis=True, label=True,pitch_color = '#538053')
fig,ax =pitch.draw(figsize =(40,20))
     
plt.scatter(pCross['Y'],pCross['X'],c='#BE2AF3',s = 350,marker = "^")


plt.scatter(pThrough['Y'],pThrough['X'],c='#60D8F6',s = 350 ,marker = "X")

plt.scatter(pWin['Y'],pWin['X'],c='#F6BF60',s = 350 ,marker = '*')

for i in range(len(pSty)):
    if(pSty[i] == 'cross'):
        plt.arrow(pDrty[i],pDrtx[i],pDesy[i]-pDrty[i],pDesx[i]-pDrtx[i], color = '#FF69B4',linewidth =4,head_width = 0.6 ,
                  length_includes_head = 'True')
    if(pSty[i] == 'through'):
        plt.arrow(pDrty[i],pDrtx[i],pDesy[i]-pDrty[i],pDesx[i]-pDrtx[i], color = '#000000',linewidth =4,head_width = 0.6 ,
                  length_includes_head = 'True')
```

```python
cross1,through1 = 0,0
for i in range (len(pSty)):
    if(pSty[i] == 'cross'):
        cross1 += 1
    if(pSty[i] == 'through'):
        through1 +=1
```

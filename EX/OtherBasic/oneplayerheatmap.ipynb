```python
from mplsoccer import Pitch,VerticalPitch
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
```

```python
data = pd.read_csv('hm.csv')
```

```python
data
```

```python
data['X'] = data['X']*1.2
data['Y'] = data['Y']*.8
```

```python
player = "suphanan"
pdata = data[data['Player'] == player ]
```

```python
pdatadef = pdata[(pdata['Event'] == 'intercept') | (pdata['Event'] == 'block') |(pdata['Event'] == 'clear') | (pdata['Event'] == 'getball')]
```

```python
pdatadef
```

```python
pdata
```

```python
pitch = VerticalPitch(pitch_length=100, pitch_width=100,
              axis=True, label=True,pitch_color = '#538053')
fig,ax =pitch.draw(figsize =(40,20))
plt.scatter(pdata['Y'],pdata['X'],s = 250)
```

```python
pitch = VerticalPitch(pitch_length=100, pitch_width=100,
              axis=True, label=True,pitch_color = '#538053')
fig,ax =pitch.draw(figsize =(40,20))

kde = sns.kdeplot(x = pdata['Y'], y = pdata['X'],
                  fill = True,shade_lowest=False,common_grid=False,
                  alpha=.5,thresh=.01,n_levels=50,cmap = 'magma',cbar=True)
plt.title('Samutprakan City Player_' + player + ' action',size = 30)
```

```python
pitch = VerticalPitch(pitch_length=100, pitch_width=100,
              axis=True, label=True,pitch_color = '#538053')
fig,ax =pitch.draw(figsize =(40,20))

kde = sns.kdeplot(x = pdatadef['Y'], y = pdatadef['X'],
                  fill = True,shade_lowest=False,common_grid=False,
                  alpha=.5,thresh=.01,n_levels=50,cmap = 'magma',cbar=True)
plt.title('Samutprakan City Player_' + player + ' def action',size = 30)
```

```python
from mplsoccer import Pitch,VerticalPitch
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
```

```python
datapass = pd.read_csv('pass.csv')
```

```python
datapass
```

```python
datapass['X'] = datapass['X']*1.2
datapass['Y'] = datapass['Y']*.8
datapass['X2'] = datapass['X2']*1.2
datapass['Y2'] = datapass['Y2']*.8
```

```python
datapass
```

```python
pCross = datapass[datapass['Event'] == 'cross' ]
pThrough = datapass[datapass['Event'] == 'through' ]
pWin = datapass[datapass['Event'] == 'win1-1' ]
```

```python
pDrtx = datapass['X']
pDrty = datapass['Y']
pDesx = datapass['X2']
pDesy = datapass['Y2']
pSty = datapass['Event']
pNamep = datapass['Player']
```

```python
pSty
```

```python
pitch = VerticalPitch(pitch_length=100, pitch_width=100,
              axis=True, label=True,pitch_color = '#538053')
fig,ax =pitch.draw(figsize =(40,20))
     
plt.scatter(pCross['Y'],pCross['X'],c='#BE2AF3',s = 350,marker = "^")


plt.scatter(pThrough['Y'],pThrough['X'],c='#60D8F6',s = 350 ,marker = "X")

plt.scatter(pWin['Y'],pWin['X'],c='#F6BF60',s = 350 ,marker = '*')

for i in range(len(pSty)):
    if(pSty[i] == 'cross'):
        plt.arrow(pDrty[i],pDrtx[i],pDesy[i]-pDrty[i],pDesx[i]-pDrtx[i], color = '#FF69B4',linewidth =4,head_width = 0.6 ,
                  length_includes_head = 'True')
    if(pSty[i] == 'through'):
        plt.arrow(pDrty[i],pDrtx[i],pDesy[i]-pDrty[i],pDesx[i]-pDrtx[i], color = '#000000',linewidth =4,head_width = 0.6 ,
                  length_includes_head = 'True')
plt.title('SPC2019 Match1 Passing Style',size = 40)
plt.text(60, -3, 'Through',c='#000000',size = 20)
plt.text(72, -3, 'Cross',c='#FF69B4',size = 20)
```

```python
player01 = 'suphanan'
pyP = datapass[datapass['Player']== player01]
ppThrough = pyP[pyP['Event'] == 'cross' ]
ppCross = pyP[pyP['Event'] == 'through' ]
ppWin = pyP[pyP['Event'] == 'win1-1' ]
ppDrtx = pyP['X']
ppDrty = pyP['Y']
ppDesx = pyP['X2']
ppDesy = pyP['Y2']
ppSty = pyP['Event']
```

```python
pitch = VerticalPitch(pitch_length=100, pitch_width=100,axis=True, label=True,pitch_color = '#538053')
fig,ax =pitch.draw(figsize =(40,20))

plt.scatter(ppCross['Y'],ppCross['X'],c='#BE2AF3',s = 350,marker = "^")


plt.scatter(ppThrough['Y'],ppThrough['X'],c='#60D8F6',s = 350 ,marker = "X")

plt.scatter(ppWin['Y'],ppWin['X'],c='#F6BF60',s = 350 ,marker = '*')

for i in range(len(pSty)):
    if(pSty[i] == 'cross'):
        if(pNamep[i] == player01):
            plt.arrow(pDrty[i],pDrtx[i],pDesy[i]-pDrty[i],pDesx[i]-pDrtx[i], color = '#FF69B4',linewidth =4,head_width = 0.6 ,
                  length_includes_head = 'True')
    if(pSty[i] == 'through'):
        if(pNamep[i] == player01):
            plt.arrow(pDrty[i],pDrtx[i],pDesy[i]-pDrty[i],pDesx[i]-pDrtx[i], color = '#000000',linewidth =4,head_width = 0.6 ,
                  length_includes_head = 'True')

plt.title('SPC2019 Match1 '+player01+' Passing Style',size = 40)
```
